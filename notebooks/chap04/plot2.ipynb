```python
#p.274
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x, dy=0.8):
    return np.sin(x) + dy * np.random.randn(50)

x = np.linspace(0, 10, 50)
y = f(x)

plt.errorbar(x, y, yerr=dy, fmt='.k', capsize=5)
```

```python
x = np.linspace(1, 10, 100)
y = np.linspace(1, 10, 100)
print(x, y)

X, Y = np.meshgrid(x, y)
print("X :", X)
print("Y :", Y)
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(X, Y) # x,y 개수?가 같아야 높이를 줘서 등고선 그릴 수 있다?
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y*x) * np.cos(X)

Z = f(X, Y)

plt.contour(X, Y, Z, 10, map='viridis')
plt.colorbar()
```

```python
plt.imshow(Z, cmap='RdGy')
```

```python
#p.288
def f(x):
    return np.sin(x)

def f2(x):
    return np.cos(x)

x = np.linspace(-10, 10, 100)

plt.plot(x, f(x), label='sin')
plt.plot(x, f2(x), label='cos')
plt.legend(loc='upper left', frameon=False, ncols=2)
```

```python
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x1 = np.linspace(-6, 6, 30)
x2 = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x1, x2)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
```
